## 9.Display Daily Sales by Day of the Week

```python
dw_mapping={
    0:'Mon',
    1:'Tues',
    2:'Wed',
    3:'Thurs',
    4:'Fri',
    5:'Sat',
    6:'Sun'
}
```

```python

data['Date'] = pd.to_datetime(data['Date'])
data['day_of_week'] = data['Date'].dt.dayofweek.map(dw_mapping)
data.head(1)
```

```python
data['day_of_week'].value_counts().plot(kind='bar')
```

## 10. Which will be the Highest Months for Sales

```python
month_mapping={
    0:'January',
    1:'February',
    2:'March',
    3:'April',
    4:'May',
    5:'June',
    6:'July',
    7:'August',
    8:'September',
    9:'October',
    10:'November',
    11:'December'
                }
```

```python
data['Month']=data['Date'].dt.month.map(month_mapping)
```
